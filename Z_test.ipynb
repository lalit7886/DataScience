## Q.N 1

Imagine you work for an e-commerce company and your team is responsible for analyzing customer purchase data.  
You want to determine whether a new website design has led to a significant increase in the average purchase amount compared to the old one.

### Old Design Data

``` python
old_design_data = [
    45.2, 42.8, 38.9, 43.5, 41.0, 44.6, 40.5, 42.6, 39.8, 41.4,
    44.3, 39.7, 42.1, 40.6, 43.0, 42.2, 41.5, 39.6, 44.0, 43.1,
    38.7, 43.9, 42.0, 41.9, 42.8, 43.7, 41.3, 40.9, 42.5, 41.6
]
```
### New Design Data
``` python
new_design_data = [
    48.5, 49.1, 50.2, 47.8, 48.7, 49.9, 48.0, 50.5, 49.8, 49.6,
    48.2, 48.9, 49.7, 50.3, 49.4, 50.1, 48.6, 58.3, 49.0, 50.0,
    48.4, 49.3, 49.5, 48.8, 50.6, 50.4, 48.1, 49.2, 50.7, 50.8
]

````python
Hn= "The new website doesnt leads to singnificant increase in sales"
Ha = "The new website design leads significant increase in sales"
alpha=0.05
poplation_std=2.5
old_design_data=np.array([45.2,42.8,38.9,43.5,41.0,44.6,40.5,42.6,39.8,41.4,44.3,39.7,42.1,40.6,43.0,42.2,41.5,39.6,44.0,43.1,38.7,43.9,42,41.9,42.8,43.7,41.3,40.9,42.5,41.6])
new_design_data=np.array([48.5,49.1,50.2,47.8,48.7,49.9,48.0,50.5,49.8,49.6,48.2,48.9,49.7,50.3,49.4,50.1,48.6,58.3,49.0,50,48.4,49.3,49.5,48.8,50.6,50.4,48.1,49.2,50.7,50.8])
n_sample=len(new_design_data)
old_mean=np.mean(old_design_data)
new_mean=np.mean(new_design_data)
n_sample=len(new_design_data)
````

````python
Z_calc=(new_mean-old_mean)/(poplation_std/np.sqrt(n_sample))
Z_table=st.norm.ppf(1-alpha)
````

````python
if Z_calc>Z_table:
    print(f"alternative hypothesis is correct: {Ha}")
    
else:
    print(f"Null hyposthesis is correct: {Hn}")
````
