```python
pop_data=[np.random.randint(1,100) for i in range (1000)]
print(pop_data[1:9])
pop_table=pd.DataFrame({"Numbers" : pop_data})
```

```python
plt.Figure(figsize=(6,4))
sns.kdeplot(x=pop_data,data=pop_table)
plt.show()
```

```python
sample_mean=[]
for no_samples in range(60):
    sample=[]
    for data in range (500):
        sample.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample))
    
sample_data= pd.DataFrame({"Mean_of_sample":sample_mean})
sns.kdeplot(x=sample_mean, data=sample_data)
```

this follow central limit theorem the distirbution of the means of the randomly taken sample follows normal distribution
